# 1. Data Processing

```python
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Função para carregar os dados do arquivo pickle
def load_data(file_path):
    """
    Carrega os dados de um arquivo pickle.
    """
    with open(file_path, 'rb') as f:
        data = pickle.load(f)
    return data

# Função para transformar a estrutura hierárquica em um DataFrame plano
def flatten_data(data):
    """
    Transforma os dados hierárquicos em um DataFrame plano.
    """
    flattened_data = []
    for syndrome_id, subjects in data.items():
        for subject_id, images in subjects.items():
            for image_id, embedding in images.items():
                flattened_data.append({
                    "syndrome_id": syndrome_id,
                    "subject_id": subject_id,
                    "image_id": image_id,
                    "embedding": embedding
                })
    return pd.DataFrame(flattened_data)

# Função para verificar embeddings e tratar inconsistências
def validate_embeddings(df):
    """
    Verifica a integridade dos embeddings e remove entradas inconsistentes.
    """
    # Verifica se todos os embeddings têm 320 dimensões
    df['embedding_length'] = df['embedding'].apply(lambda x: len(x))
    inconsistent_embeddings = df[df['embedding_length'] != 320]
    if not inconsistent_embeddings.empty:
        print("Inconsistências detectadas nos embeddings:")
        print(inconsistent_embeddings)
        # Remover embeddings inconsistentes
        df = df[df['embedding_length'] == 320]
    df.drop(columns=['embedding_length'], inplace=True)
    return df

# Função para verificar integridade e possíveis problemas nos dados
def check_data_integrity(df):
    """
    Exibe informações gerais e verifica dados ausentes.
    """
    print("Informações gerais do DataFrame:")
    print(df.info())
    print("\nVerificando valores ausentes:")
    print(df.isnull().sum())
    print("\nExemplo de dados:")
    print(df.head())

# Função para análise de distribuição de embeddings
def embedding_statistics(df):
    """
    Calcula estatísticas e exibe a distribuição dos valores nos embeddings.
    """
    embeddings_array = np.array(df['embedding'].tolist())
    print("Estatísticas gerais dos embeddings:")
    print("Média:", np.mean(embeddings_array))
    print("Desvio padrão:", np.std(embeddings_array))
    print("Min:", np.min(embeddings_array))
    print("Max:", np.max(embeddings_array))

    # Visualização da distribuição
    plt.figure(figsize=(10, 6))
    plt.hist(embeddings_array.flatten(), bins=50, alpha=0.7)
    plt.title('Distribuição de Valores nos Embeddings')
    plt.xlabel('Valor')
    plt.ylabel('Frequência')
    
plt.savefig('grafico_gerado.png', dpi=300, bbox_inches='tight')
plt.show()

# Função para análise exploratória dos dados
def exploratory_data_analysis(df):
    """
    Realiza análise exploratória básica dos dados.
    """
    print("Número total de imagens:", len(df))
    print("Número total de síndromes únicas:", df['syndrome_id'].nunique())
    
    syndrome_counts = df['syndrome_id'].value_counts()
    print("\nDistribuição de imagens por síndrome:")
    print(syndrome_counts)
    
    # Visualização da distribuição das síndromes
    plt.figure(figsize=(10, 6))
    syndrome_counts.plot(kind='bar')
    plt.title('Distribuição de Imagens por Síndrome')
    plt.xlabel('Síndrome ID')
    plt.ylabel('Quantidade de Imagens')
    
plt.savefig('grafico_gerado.png', dpi=300, bbox_inches='tight')
plt.show()

# Caminho para o arquivo pickle
file_path = 'mini_gm_public_v0.1.p'

# Pipeline de processamento
data = load_data(file_path)
df = flatten_data(data)
df = validate_embeddings(df)
```

# 2. Data Visualization

```python
from sklearn.manifold import TSNE
import seaborn as sns

# Função para aplicar t-SNE e plotar os embeddings em 2D
def visualize_embeddings_tsne(df, perplexity=30, learning_rate=200, random_state=42):
    """
    Reduz a dimensionalidade dos embeddings para 2D usando t-SNE e visualiza os resultados.
    
    Parâmetros:
        df (DataFrame): DataFrame contendo os embeddings e síndrome_id.
        perplexity (int): Parâmetro de perplexidade do t-SNE.
        learning_rate (float): Taxa de aprendizado para t-SNE.
        random_state (int): Seed para reprodutibilidade.
    """
    # Extrai os embeddings e os rótulos
    embeddings = np.array(df['embedding'].tolist())
    syndrome_ids = df['syndrome_id']

    # Aplica t-SNE para reduzir para 2 dimensões
    tsne = TSNE(n_components=2, perplexity=perplexity, learning_rate=learning_rate, random_state=random_state)
    embeddings_2d = tsne.fit_transform(embeddings)

    # Cria um DataFrame para os resultados reduzidos
    tsne_df = pd.DataFrame({
        'Dimension 1': embeddings_2d[:, 0],
        'Dimension 2': embeddings_2d[:, 1],
        'Syndrome ID': syndrome_ids
    })

    # Visualiza os clusters
    plt.figure(figsize=(10, 8))
    sns.scatterplot(
        data=tsne_df,
        x='Dimension 1',
        y='Dimension 2',
        hue='Syndrome ID',
        palette='tab10',
        s=50,
        alpha=0.7
    )
    plt.title('Visualização t-SNE dos Embeddings por Síndrome ID')
    plt.xlabel('Dimensão 1')
    plt.ylabel('Dimensão 2')
    plt.legend(loc='best', title='Síndrome ID', bbox_to_anchor=(1.05, 1), borderaxespad=0.)
    plt.tight_layout()
    
plt.savefig('grafico_gerado.png', dpi=300, bbox_inches='tight')
plt.show()

# Chamada da função de visualização
visualize_embeddings_tsne(df)
```

A análise da visualização gerada pelo t-SNE revela importantes insights sobre os embeddings e sua relação com as síndromes. Algumas síndromes, como 30000034 (amarelo) e 30000082 (azul), apresentam clusters relativamente bem definidos, indicando que seus embeddings possuem separabilidade no espaço de 2D, o que pode facilitar a tarefa de classificação. Por outro lado, outras síndromes, como 700018215 (verde) e 100180860 (cinza), mostram maior dispersão e sobreposição com outras classes, o que pode dificultar o aprendizado do modelo.

A sobreposição entre as classes sugere que os embeddings de algumas síndromes não são suficientemente discriminados, o que pode levar a confusões durante a classificação. Isso é particularmente evidente em síndromes como 100180860, que compartilham áreas densas no gráfico com outras classes adjacentes.

Esses padrões têm implicações diretas para a tarefa de classificação. Síndromes que formam clusters bem definidos podem ser classificadas com maior precisão, enquanto aquelas com maior dispersão ou sobreposição provavelmente impactarão negativamente métricas como a precisão e o F1-Score. Para lidar com esses desafios, pode ser necessário explorar modelos mais sofisticados, como SVMs com kernels não lineares ou redes neurais profundas, que têm maior capacidade de capturar relações complexas entre os dados. Métodos como aumento de dados (data augmentation) ou ajustes nos embeddings pré-treinados também podem ajudar a melhorar a separação entre as classes.

Em resumo, o t-SNE oferece uma visão clara da estrutura dos embeddings, destacando tanto as classes mais fáceis de distinguir quanto aquelas que apresentam desafios significativos para o modelo de classificação. Esses resultados devem guiar estratégias de modelagem e processamento adicional para melhorar o desempenho geral do modelo.

# 3. Classification Task

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_auc_score, f1_score, accuracy_score, top_k_accuracy_score
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Função para calcular métricas
def evaluate_model(y_true, y_pred, y_prob, k):
    """
    Avalia o desempenho do modelo com base em diferentes métricas.
    """
    auc = roc_auc_score(y_true, y_prob, multi_class='ovr')
    f1 = f1_score(y_true, y_pred, average='weighted')
    top_k_acc = top_k_accuracy_score(y_true, y_prob, k=k)
    return auc, f1, top_k_acc

# Função para realizar KNN com Cross-Validation
def knn_classification(df, k_values=range(1, 16), metric='euclidean'):
    """
    Realiza classificação KNN com validação cruzada e retorna métricas de desempenho.
    """
    embeddings = np.array(df['embedding'].tolist())
    labels = df['syndrome_id']
    skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

    results = {k: {"AUC": [], "F1-Score": [], "Top-1 Accuracy": []} for k in k_values}

    for train_index, test_index in skf.split(embeddings, labels):
        X_train, X_test = embeddings[train_index], embeddings[test_index]
        y_train, y_test = labels.iloc[train_index], labels.iloc[test_index]
        
        for k in k_values:
            # Define o modelo KNN
            knn = KNeighborsClassifier(n_neighbors=k, metric=metric)
            knn.fit(X_train, y_train)
            
            # Predições
            y_pred = knn.predict(X_test)
            y_prob = knn.predict_proba(X_test)
            
            # Avaliação do modelo
            auc, f1, top_k_acc = evaluate_model(y_test, y_pred, y_prob, k=1)
            results[k]["AUC"].append(auc)
            results[k]["F1-Score"].append(f1)
            results[k]["Top-1 Accuracy"].append(top_k_acc)

    # Agrega os resultados
    aggregated_results = {k: {metric: np.mean(values[metric]) for metric in values} for k, values in results.items()}
    return aggregated_results

# Função para comparar métricas entre as duas distâncias
def compare_metrics(df, k_values=range(1, 16)):
    """
    Compara as métricas de classificação para as distâncias Euclidean e Cosine.
    """
    print("Calculando para Métrica Euclidean...")
    results_euclidean = knn_classification(df, k_values=k_values, metric='euclidean')
    
    print("Calculando para Métrica Cosine...")
    results_cosine = knn_classification(df, k_values=k_values, metric='cosine')

    return results_euclidean, results_cosine

# Executa o KNN e compara as distâncias
k_values = range(1, 16)
results_euclidean, results_cosine = compare_metrics(df, k_values=k_values)

# Converte os resultados em DataFrames
euclidean_df = pd.DataFrame(results_euclidean).T
cosine_df = pd.DataFrame(results_cosine).T

# Exibe os resultados
print("\nResultados - Métrica Euclidean")
print(euclidean_df)

print("\nResultados - Métrica Cosine")
print(cosine_df)

# Visualização das Métricas
def plot_results(results, title):
    """
    Plota os resultados das métricas para diferentes valores de k.
    """
    ks = list(results.keys())
    auc = [results[k]["AUC"] for k in ks]
    f1 = [results[k]["F1-Score"] for k in ks]
    top1 = [results[k]["Top-1 Accuracy"] for k in ks]

    plt.figure(figsize=(10, 6))
    plt.plot(ks, auc, label="AUC")
    plt.plot(ks, f1, label="F1-Score")
    plt.plot(ks, top1, label="Top-1 Accuracy")
    plt.title(title)
    plt.xlabel("Número de Vizinhos (k)")
    plt.ylabel("Métrica")
    plt.legend()
    plt.grid()
    
plt.savefig('grafico_gerado.png', dpi=300, bbox_inches='tight')
plt.show()

# Plota os resultados para Euclidean
plot_results(results_euclidean, "Desempenho - Métrica Euclidean")

# Plota os resultados para Cosine
plot_results(results_cosine, "Desempenho - Métrica Cosine")
```

A análise dos resultados mostra que a métrica Euclidean apresenta desempenho superior para valores mais altos de k, especialmente em termos de AUC e Top-1 Accuracy, enquanto a métrica Cosine se destaca para valores menores de k, com melhor equilíbrio entre precisão e recall (F1-Score). A Euclidean continua a melhorar com o aumento de k, indicando maior eficácia na captura de relações globais entre os embeddings. Em contraste, a Cosine estabiliza mais cedo, sugerindo que ela é mais eficaz em capturar padrões locais e lidar com distribuições sobrepostas.

As diferenças de desempenho são explicadas pela natureza das métricas: a Euclidean mede a distância linear, sendo mais sensível à separação global dos dados, enquanto a Cosine mede similaridade angular, sendo mais robusta para diferenças de magnitude nos vetores. Assim, a escolha da métrica deve considerar o objetivo da tarefa e o valor de k: Euclidean para maior estabilidade em k altos e Cosine para granularidade em k baixos.

# 4. Metrics and Evaluation

```python
from sklearn.metrics import roc_curve, auc

# Função para calcular e plotar curvas ROC AUC médias
def plot_roc_curves(df, k, metric, ax):
    """
    Calcula e plota a curva ROC AUC média para a métrica especificada.
    """
    embeddings = np.array(df['embedding'].tolist())
    labels = pd.get_dummies(df['syndrome_id']).values  # Codificação one-hot
    skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

    mean_fpr = np.linspace(0, 1, 100)
    tprs = []
    aucs = []

    for train_index, test_index in skf.split(embeddings, df['syndrome_id']):
        X_train, X_test = embeddings[train_index], embeddings[test_index]
        y_train, y_test = labels[train_index], labels[test_index]

        # Modelo KNN
        knn = KNeighborsClassifier(n_neighbors=k, metric=metric)
        knn.fit(X_train, np.argmax(y_train, axis=1))
        y_prob = knn.predict_proba(X_test)

        # Curvas ROC para cada classe
        for i in range(y_test.shape[1]):
            fpr, tpr, _ = roc_curve(y_test[:, i], y_prob[:, i])
            interp_tpr = np.interp(mean_fpr, fpr, tpr)
            interp_tpr[0] = 0.0
            tprs.append(interp_tpr)
            roc_auc = auc(fpr, tpr)
            aucs.append(roc_auc)

    # Média e intervalo da curva ROC
    mean_tpr = np.mean(tprs, axis=0)
    mean_tpr[-1] = 1.0
    mean_auc = auc(mean_fpr, mean_tpr)

    # Plotagem
    ax.plot(mean_fpr, mean_tpr, label=f'{metric.capitalize()} (AUC = {mean_auc:.2f})')
    ax.fill_between(
        mean_fpr,
        np.maximum(mean_tpr - np.std(tprs, axis=0), 0),
        np.minimum(mean_tpr + np.std(tprs, axis=0), 1),
        alpha=0.2,
    )
    return mean_auc

# Plotando as curvas para Cosine e Euclidean
fig, ax = plt.subplots(figsize=(10, 6))
auc_euclidean = plot_roc_curves(df, k=5, metric='euclidean', ax=ax)
auc_cosine = plot_roc_curves(df, k=5, metric='cosine', ax=ax)

# Configuração do gráfico
ax.plot([0, 1], [0, 1], 'k--', lw=2)
ax.set_title('Curvas ROC AUC Médias - Euclidean vs Cosine')
ax.set_xlabel('Taxa de Falsos Positivos (FPR)')
ax.set_ylabel('Taxa de Verdadeiros Positivos (TPR)')
ax.legend()
plt.grid()
plt.tight_layout()

plt.savefig('grafico_gerado.png', dpi=300, bbox_inches='tight')
plt.show()

# Exibição das tabelas de métricas
print("\nTabela de Resultados - Métrica Euclidean")
print(euclidean_df)

print("\nTabela de Resultados - Métrica Cosine")
print(cosine_df)
```

As curvas ROC AUC do gráfico acima mostram que ambas as métricas apresentam desempenhos excelentes para a classificação de síndromes. A métrica Cosine alcançou uma AUC média de 0.93, enquanto a métrica Euclidean apresentou uma AUC de 0.91. A curva da métrica Cosine está ligeiramente mais próxima do canto superior esquerdo, indicando melhor separação entre as classes. Além disso, a região sombreada, que representa a variação entre os folds de validação cruzada, é menor para a métrica Cosine, sugerindo maior consistência nos resultados, enquanto a métrica Euclidean apresenta maior variabilidade em alguns pontos. Essa diferença pode ser explicada pela robustez da métrica Cosine em capturar similaridades angulares entre os vetores, especialmente útil quando os embeddings apresentam variações em suas magnitudes. Por outro lado, a métrica Euclidean mede distâncias lineares, sendo mais eficaz para capturar distribuições globais no espaço de alta dimensionalidade.

```python
# Plota os resultados para Euclidean
plot_results(results_euclidean, "Desempenho - Métrica Euclidean")

# Plota os resultados para Cosine
plot_results(results_cosine, "Desempenho - Métrica Cosine")
```

Nos gráficos de desempenho para diferentes valores de k acima, observa-se que a AUC é consistentemente alta para ambas as métricas. Na métrica Euclidean, a AUC cresce rapidamente com o aumento de k, estabilizando acima de 0.95 a partir de k=7. Já a métrica Cosine apresenta um crescimento mais rápido nos primeiros valores de k, estabilizando em torno de 0.93 entre k=6 e k=10. No caso do F1-Score, que reflete o equilíbrio entre precisão e recall, a métrica Cosine apresenta melhor desempenho para valores de k menores ou iguais a 5, indicando maior eficácia em granularidades mais finas e em cenários onde as classes possuem maior sobreposição. Entretanto, para valores de k maiores que 5, a métrica Euclidean supera a Cosine, mostrando maior estabilidade. Em relação à precisão Top-1, ambas as métricas apresentam desempenhos similares para valores baixos de k. No entanto, a partir de k=6, a Euclidean demonstra uma leve superioridade, refletindo maior capacidade de capturar padrões globais ao considerar mais vizinhos.

De maneira geral, a métrica Cosine destaca-se para valores baixos de k, graças à sua capacidade de capturar similaridades locais entre os embeddings, o que é particularmente útil em cenários onde as classes possuem maior sobreposição. Por outro lado, a métrica Euclidean apresenta melhor desempenho em valores altos de k, beneficiando-se de sua capacidade de capturar padrões globais ao considerar mais vizinhos. Ambas as métricas são robustas e consistentes em termos de desempenho, sendo adequadas para a tarefa de classificação de síndromes a partir dos embeddings. A escolha entre as métricas deve considerar as características específicas do problema. A métrica Cosine pode ser preferida em conjuntos de dados com maior variação de magnitude nos embeddings ou em casos onde há maior sobreposição entre classes, enquanto a métrica Euclidean é vantajosa em cenários que exigem maior estabilidade e capacidade de capturar estruturas globais.
